```python
import zipfile
with zipfile.ZipFile('2021_GCP_all_for_AUS_short-header.zip', 'r') as zip_ref:
    zip_ref.extractall('./unzip')
```

```python
# df = pd.read_csv('./unzip/2021 Census GCP All Geographies for AUS/*/*')

import pandas as pd
import glob

# Specify the directory and file pattern
directory_path = './unzip/2021 Census GCP All Geographies for AUS/*/'
file_pattern = '*.csv'

# Use glob to get a list of file paths matching the pattern
file_paths = glob.glob(directory_path + file_pattern)

# Initialize an empty DataFrame to concatenate the data
df = pd.DataFrame()

# Loop through the list of file paths and read each CSV file
for file_path in file_paths:
    # Read the CSV file and concatenate it to the DataFrame
    df_temp = pd.read_csv(file_path)
    df = pd.concat([df, df_temp], ignore_index=True)

# Now, 'df' contains the concatenated data from all CSV files in the specified directory
```

```python
# Assuming 'M_15_19_CF_Own_CCO' is the column name
column_name = 'M_15_19_CF_Own_CCO'

# Filter rows where the specified column is not NaN
df_filtered = df[df[column_name].notna()]

# Now, df_filtered contains only the rows where 'M_15_19_CF_Own_CCO' is not NaN
```
